```python
import os
import zipfile
import pandas as pd
import regex as re
```

```python
with open('Dataset.txt', 'r') as file:
    text = file.read()
    print(text)
```

```python
text = text.split(' ')
text
```

```python
text = [i.lower() for i in text]
print(text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
otak = TfidfVectorizer()
otak.fit(text)
```

```python
y = otak.transform(text)
y = y.toarray()
y
```

```python
otak.get_feature_names()
```

```python
cari = input('Masukkan yang anda cari: ')
v_cari = otak.transform ([cari]).toarray()
indeks = 0 
hasil = 0

from sklearn.metrics.pairwise import cosine_similarity
for i, v in enumerate (y):
  mirip = cosine_similarity ([v],v_cari)
  if mirip > hasil:
    hasil = mirip
    indeks = i
    
print(text[indeks])
```

```python
from Levenshtein import distance

def hitung_jarak(kata1, kata2):
    return distance(kata1, kata2)
```

```python
def search_engine(keyword, text):
    jarak_terdekat = float("inf")
    hasil_cocok = ""
    for data in text:
        jarak = hitung_jarak(keyword, data)
        if jarak < jarak_terdekat:
            jarak_terdekat = jarak
            hasil_cocok = data
    return hasil_cocok

keyword = input("Masukkan Kalimat: ")
hasil = search_engine(keyword, text)
print("Mungkin maksud anda: ", (hasil))
```
